### **Dados de Restaurantes do iFood (fevereiro de 2021)**

Sobre o conjunto de dados: 
+ O código de 7 dígitos do IBGE é utilizado para encontrar o restaurante ao fazer a raspagem de dados, no entanto, não é representativo da cidade em que o restaurante está localizado. Para obter a localização exata, optamos por extrai-la do restaurante a partir da URL indicada.

Importando bibliotecas e carregando o arquivo em um dataframe

```python
import pandas as pd

df = pd.read_csv('ifood-restaurants-february-2021.csv')
df
```

Descrição dos dados e alguns tratamentos

```python
# Visualização de informações do dataframe
df.info()
```

```python
# Visualização de valores nulos
nulos_AFS = df['availableForScheduling'].isnull().sum()
nulos_avatar = df['avatar'].isnull().sum()
nulos_category = df['category'].isnull().sum()
nulos_delivery_fee = df['delivery_fee'].isnull().sum()
nulos_delivery_time = df['delivery_time'].isnull().sum()
nulos_distance = df['distance'].isnull().sum()
nulos_ibge = df['ibge'].isnull().sum()
nulos_minimumOrderValue = df['minimumOrderValue'].isnull().sum()
nulos_name = df['name'].isnull().sum()
nulos_paymentCodes = df['paymentCodes'].isnull().sum()
nulos_rating = df['rating'].isnull().sum()
nulos_tags = df['tags'].isnull().sum()
nulos_url = df['url'].isnull().sum()


print(f'availableForScheduling: {nulos_AFS} nulos')
print(f'avatar: {nulos_avatar} nulos')
print(f'category: {nulos_category} nulos')
print(f'delivery_fee: {nulos_delivery_fee} nulos')
print(f'delivery_time: {nulos_delivery_time} nulos')
print(f'distance: {nulos_distance} nulos')
print(f'ibge: {nulos_ibge} nulos')
print(f'minimumOrderValue: {nulos_minimumOrderValue} nulos')
print(f'name: {nulos_name} nulos')
print(f'paymentCodes: {nulos_paymentCodes} nulos')
print(f'rating: {nulos_rating} nulos')
print(f'tags: {nulos_tags} nulos')
print(f'url: {nulos_url} nulos')
```

```python
# Visualização das colunas
df.columns
```

```python
# Exclusão de algumas colunas
excluir_colunas = ['availableForScheduling','avatar', 'tags','ibge' ,'paymentCodes']
df = df.drop(excluir_colunas, axis=1)
```

```python
df.head()
```

```python
# Extração da localização a partir da URL dos restaurantes
df['localization'] = ''

for index, row in df.iterrows():
    url = row['url']
    df.at[index, 'localization'] = url[34:].split('/')[0]

excluir_coluna = ['url']
df = df.drop(excluir_coluna, axis=1)

df.head()
```

```python
# Separação de Cidade e Estado
df['state'] = ''
df['city'] = ''

for index, row in df.iterrows():
    localization = row['localization'] 
    estado = localization[-2:].upper()
    cidade = localization[:-2].replace('-', ' ').title()
    
    df.at[index, 'state'] = estado
    df.at[index, 'city'] = cidade

excluir_coluna = ['localization']
df = df.drop(excluir_coluna, axis=1)

df.head()
```

```python
# Transforma variáveis string em números, para que possam ser manipuladas
# Float para valores, int para minutos, int para kilômetros
df.fillna(0, inplace=True)

df['delivery_fee'] = df['delivery_fee'].replace('[\\$,]', '', regex=True).astype(float)
df['minimumOrderValue'] = df['minimumOrderValue'].replace('[\\$,]', '', regex=True).astype(float)

df['distance'] = df['distance'].astype(int)
df['delivery_time'] = df['delivery_time'].astype(int)

df.head()
```

```python
# Retirando valores acima de 10000 na coluna minimumOrderValue, por ser muito fora da curva
df = df.loc[df["minimumOrderValue"] < 10000]
df = df.loc[df["delivery_time"] < 999]
```

```python
contagem = df.groupby("category").size().reset_index(name='contagem').sort_values("contagem", ascending=False)
contagem
```

Importando bibliotecas e preparando plotagem de gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Top 5 de categorias mais pedidas

# Configuração do estilo do eixo
with sns.axes_style('whitegrid'):
    grafico = sns.barplot(data=contagem.head(5), x="category", y="contagem", hue="category", palette="pastel")
    
    for p in grafico.patches:
        grafico.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)
    
    grafico.set_title('Top 5 categorias mais pedidas', fontsize=16)
    grafico.set_xlabel('Categoria', fontsize=14)
    grafico.set_ylabel('Quantidade', fontsize=14)
    grafico.figure.set_size_inches(w=20/2.54, h=10/2.54)
```

```python
# Top 5 de categorias menos pedidas

with sns.axes_style('whitegrid'):
    grafico = sns.barplot(data=contagem.tail(5), x="category", y="contagem", hue="category", palette="pastel")
    
    for p in grafico.patches:
        grafico.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)
    
    grafico.set_title('Top 5 categorias menos pedidas', fontsize=16)
    grafico.set_xlabel('Categoria', fontsize=14)
    grafico.set_ylabel('Quantidade', fontsize=14)
    grafico.figure.set_size_inches(w=20/2.54, h=10/2.54)
```

```python
# Taxa de entrega x Distância

with sns.axes_style("whitegrid"):
    grafico = sns.scatterplot(data=df, x="delivery_fee", y="distance", hue="delivery_fee")
    grafico.set_title('Relação entre Taxa de Entrega e Distância', fontsize=16)
    grafico.set_xlabel('Taxa de Entrega (R$)', fontsize=14)
    grafico.set_ylabel('Distância (metros)', fontsize=14)
    grafico.figure.set_size_inches(w=10, h=6)
    
    plt.show()
```

```python
# Faixa de preço x Avaliação

avg_rating_by_price = df.groupby('price_range')['rating'].mean().reset_index()

with sns.axes_style("whitegrid"):
    grafico = sns.barplot(data=avg_rating_by_price, x="price_range", y="rating", hue="price_range", palette="pastel")
    grafico.set_title('Média de Avaliação por Faixa de Preço', fontsize=16)
    grafico.set_xlabel('Faixa de Preço', fontsize=14)
    grafico.set_ylabel('Avaliação Média', fontsize=14)
    grafico.set_ylim(0, 5)
    grafico.figure.set_size_inches(w=10, h=6)
    
    plt.show()
```

```python
# Top 10 cidades que mais consomem iFood
top_10_cidades = df['city'].value_counts().head(10).index
df_top_10_cidades = df[df['city'].isin(top_10_cidades)].copy()

with sns.axes_style("whitegrid"):
    grafico = sns.barplot(data=df_top_10_cidades[df_top_10_cidades['minimumOrderValue'] <= 100], x="minimumOrderValue", y="city", palette="pastel")
    grafico.set_title('Valores mínimos de pedido nas 10 cidades que mais pedem', fontsize=16)
    grafico.set_xlabel('Valor mínimo do pedido (R$)', fontsize=14)
    grafico.set_ylabel('Cidades', fontsize=14)
    grafico.figure.set_size_inches(w=10, h=6)
    
    plt.show()
```

```python
# Preço de entrega por Estado 

with sns.axes_style("whitegrid"):
    grafico = sns.barplot(data=df, x="state", y="delivery_fee", hue="state", palette="pastel")
    grafico.set_title('Preço de entrega por estado', fontsize=16)
    grafico.set_xlabel('Estado', fontsize=14)
    grafico.set_ylabel('Valor da entrega (R$)', fontsize=14)
    grafico.figure.set_size_inches(w=15, h=6)

    plt.show()
```
